First start with all the libraries needed to perform Exploratory Data Analysis

```python
import pandas
import seaborn
import matplotlib.pyplot as plt
```

Now we can read the data into pandas dataframe

```python
data = pandas.read_csv("../data/winequality-red.csv")
```

So now we have loaded the data in the form of dataframe, lets look at it

```python
data.head(5)
```

The above head() function show top-5 rows of the data. If we want to see bottom-5 rows, we can use tail(). We can also pass any number in these functions to get predefined rows. For example

```python
data.tail(10)
```

Now let's take a closer look at the data

```python
data.shape
```

This data has 1599 rows (excluding headers) and 12 coloumns

It is also a good practice to know the columns and their corresponding data types,along with finding whether they contain null values or not.

```python
data.info()
```

All the columns are float and does not contain any null value. Only the quality is integer.

The describe() function in pandas is very handy in getting various summary statistics.This function returns the count, mean, standard deviation, minimum and maximum values and the quantiles of the data.

```python
data.describe()
```

1. Here as you can notice mean value is less than median value of each column which is represented by 50%(50th percentile) in index column.
2. There is notably a large difference between 75th %tile and max values of predictors “residual sugar”,”free sulfur dioxide”,”total sulfur dioxide”.
3. Thus observations 1 and 2 suggests that there are extreme values-Outliers in our data set.

Now let's look at our only integer variable quality

```python
data.quality.unique()
```

1. Target variable/Dependent variable is discrete and categorical in nature.
2. “quality” score scale ranges from 1 to 10;where 1 being poor and 10 being the best.
3. 1,2 & 10 Quality ratings are not given by any observation. Only scores obtained are between 3 to 9.

```python
data.quality.value_counts()
```

1. This tells us vote count of each quality score in descending order.
2. “quality” has most values concentrated in the categories 5, 6 and 7.
3. Only a few observations made for the categories 3 & 9.

Now let's see how the variables are correlated with each other

```python
data.corr()
```

We can generate a plot to make the visualization better 

```python
fig, ax = plt.subplots(figsize=(15, 15))
seaborn.heatmap(data.corr(),cmap="YlGnBu", annot=True,ax=ax)
plt.show()
```

1. Here we can infer that “density” has strong positive correlation with “fixed acidity”.
2. “free sulphur dioxide” and “pH” has almost no correlation with “quality”.
3. Since correlation is zero we can infer there is no linear relationship between these two predictors.However it is safe to drop these features in case you’re applying Linear Regression model to the dataset.
