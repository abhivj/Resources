## 7. ML Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
telecom_data.columns
```

```python
telecom_data["international plan"] = telecom_data["international plan"].map({"yes": 1, "no": 0})
telecom_data["voice mail plan"] = telecom_data["voice mail plan"].map({"yes": 1, "no": 0})
telecom_data_sample = telecom_data[['total day minutes','international plan','voice mail plan','churn']]
y = telecom_data_sample["churn"].astype("int").values
X = telecom_data_sample.drop("churn", axis=1).values
```

```python
scaled_X = StandardScaler().fit_transform(X)
```

```python
scaled_X
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.30)
```

```python
model = LogisticRegression().fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
```

```python
accuracy_score(y_test,prediction)
```
